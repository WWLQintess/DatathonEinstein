# Copiar os arquivos para subdiretórios ICDR + 0, 1, 2, 3 ou 4 conforme a classificação dada no arquivo lido acima

```python
import shutil

path_src = r'C:\Users\Resource\Resource Tecnologia e Informatica Ltda\Data Science & Analytics - General\Projetos Internos\Datathon\Einstein Frontiers\Images'
path_src += '\\'
path_dstntn = r'C:\Users\Resource\Resource Tecnologia e Informatica Ltda\Data Science & Analytics - General\Projetos Internos\Datathon\Einstein Frontiers\ICDR'

for i in range(df_einstein_lbls.shape[0]):
    path_src_arq = path_src + str(df_einstein_lbls.ID[i]) + '.jpg'
    path_dstntn_arq = path_dstntn + str(df_einstein_lbls.ICDR[i]) + '\\' + str(df_einstein_lbls.ID[i]) + '.jpg'
    shutil.copyfile(path_src_arq, path_dstntn_arq)
```
